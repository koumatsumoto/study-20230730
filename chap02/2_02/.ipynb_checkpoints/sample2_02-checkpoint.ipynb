```python
import numpy as np
import matplotlib.pyplot as plt

def standardize(x):
    '''
        標準化を行う
        x : 標準化前のx
    '''
    x_mean = x.mean()         # 平均値を求める
    std = x.std()             # 標準偏差を求める
    
    return (x - x_mean) / std # 標準化した値を返す

def create_matrix(x_std):
    '''
        全ての要素が1のベクトル、x_std、x_stdの2乗を
        縦方向に連結して行列を作成し、転置して返す
        ------------
        x_std : 標準化後のx
    '''
    return np.vstack([
        np.ones(         # 値がすべて1のベクトル
            x_std.size), # (要素数はx_stdと同じ)
        x_std,           # x_stdは標準化後のx
        x_std ** 2       # x_stdのすべての要素を2乗する
        ]
    ).T                  # 作成した行列を転置する

def f(matrix_X, parameter):
    '''
        多項式回帰
        ------------
        matrix_X  : 行列X
        parameter : パラメーターw、w1、w2の行ベクトル
    '''
    # 多項式回帰fw(x)=Xw
    return np.dot(matrix_X, parameter)

def E(matrix_X, y, parameter):
    '''
        誤差関数
        ------------
        matrix_X  : xの行列
        y         : yの値
        parameter : w0,w1,w3のベクトル
    '''
    return 0.5 * np.sum((y - f(matrix_X,
                               parameter)
                         ) ** 2)
    
def polynomial_regression(matrix_X, y):
    '''
        多項式回帰で最適化を行う
        ------------
        matrix_X : 行列X
        y        : yの値
    '''
    # パラメーターw0,w1,w2を0〜50の整数で初期化
    parameter = np.random.randint(0,50,3)
    # 学習率を0.001に設定
    LNR = 1e-3
    # 誤差関数の差分を保持する変数を1で初期化
    difference  = 1
    # 更新回数をカウントする変数を1で初期化
    count = 1
    # パラメーターの初期値に誤差関数を適用してbeforeに代入
    err_before = E(matrix_X, # 行列X
                   y,        # yの値
                   parameter # パラメーターw0,w1,w2のベクトル
                   )
    
    # ログの書式
    log = '({}) parameter: {} error: {:.4f}'  
    # 誤差の差分が0.01以下になるまでパラメータの更新を繰り返す
    while difference  > 1e-2:
        # w0,w1,w3を更新する
        parameter = parameter - LNR * np.dot(
            f(matrix_X, parameter) - y,
            matrix_X
        )

        # 更新後のw0,w1,w3の値で誤差関数の値を求める
        err_after = E(matrix_X, # 行列X
                      y,        # yの値
                      parameter # パラメーターのベクトル
                      )

        # パラメーター更新前後における誤差の差分を求める
        difference  = err_before - err_after
        # パラメーター更新後の誤差をerr_beforeの値にする
        err_before = err_after 
        # 最初の1回と以降100回ごとにログを出力
        if (count == 1 or
            count % 100 == 0):
            print(log.format(count, parameter, difference))
        # カウンター変数の値を1増やす
        count += 1
        
    # 最後の更新時のログを出力
    print(log.format(count, parameter, difference))
    # 更新終了後のw0,w1,w3の値を戻り値として返す
    return parameter
```

```python
# 学習データを読み込む
data = np.loadtxt('sales.csv',   # 読み込むファイル
                  dtype='int',   # データ型を指定
                  delimiter=',', # 区切り文字を指定
                  skiprows=1     # 1行目を読み飛ばす
                  )
x = data[:,0]           # 1列目の成分をxに代入
y = data[:,1]           # 2列目の成分をyに代入
x_std = standardize(x)  # xのすべての要素を標準化

# 標準化したxから行列Xを作成
matrix_X = create_matrix(x_std)
# 多項式回帰でパラメーターΦ0、Φ1の値を求める
parameter = polynomial_regression(matrix_X, y)
```

```python
# x軸の値として-3から3までの等差数列を生成
x_axis = np.linspace(start=-3, # 数列の始点
                     stop=3,   # 数列の終点
                     num=100   # 等差数列の要素数
                     )
# y軸の範囲を設定
plt.ylim(300, 2500)
# 学習データをプロット
plt.plot(x_std, y, 'o')
# 多項式回帰で求めた曲線をプロット
plt.plot(
    # x軸の値
    x_axis,
    # yの予測値
    f(create_matrix(x_axis), parameter)
    )
plt.grid(True)
# グラフを表示
plt.show()
```

```python
# 予測する
input_x = input('予測に使用するxの値を入力してください>')
x_mean = x.mean()                            # xの平均値を求める
std = x.std()                                # xの標準偏差を求める
pred = (int(input_x) - x_mean)/std           # 入力値を標準化する
print(parameter[0]                           # 予測したyの値を出力
      + pred*parameter[1]
      + (pred**2)*parameter[2])
```
